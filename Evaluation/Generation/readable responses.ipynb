```python
def clean_text(text):
    if isinstance(text, str):
        return (
            text.replace("\\n", "\n")  
                .replace("\\'", "'")  
                .replace('("', '') 
                .replace('")', '') 
        ).strip()  
    return text
```

```python
df["Response_Only2"] = df["Response_Only"].apply(clean_text)
```

```python
for index, row in df.iterrows():
    print(f"Query: {row['Query']}\n")
    print(f"Response Only: {row['Response_Only2']}\n")
    print("=" * 80)  # Separator for better readability
    print("\n\n")
```
